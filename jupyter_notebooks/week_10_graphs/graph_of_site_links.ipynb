```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url="http://books.toscrape.com/"
list_of_all_urls=[]
list_of_visited_urls=[]
```

```python
def get_links_from_page(url):
    result = requests.get(url)
    soup = BeautifulSoup(result.content)
    list_of_links=[]
    for link in soup.find_all('a', href=True):
        list_of_links.append(link['href'])
    return list_of_links
```

```python
reslts_list=get_links_from_page(base_url)
```

```python
for this_url in reslts_list:
    list_of_all_urls.append(base_url+this_url)

list_of_all_urls=list(set(list_of_all_urls))
```
